```python
df = pd.read_csv('final_dataset.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Ensure 'DoW' and 'Title' remain as object (string) types
df['DoW'] = df['DoW'].astype('object')
df['Title'] = df['Title'].astype('object')

# Convert 'Month', 'Year', 'IS_WEEKDAY', and 'IS_HOLIDAY' to integers
df['Month'] = df['Month'].astype('int')
df['Year'] = df['Year'].astype('int')
df['IS_HOLIDAY'] = df['IS_HOLIDAY'].fillna(0).astype('int')  # Handling NaN values if any
df['IS_WEEKDAY'] = df['IS_WEEKDAY'].astype('int')

# Convert all open, close, high, and low columns to float, handling object columns first
float_cols = ['btc_open', 'btc_high', 'btc_low', 'btc_close', 
              'oil_open', 'oil_close', 'oil_high', 'oil_low', 
              'dxy_open', 'dxy_close', 'dxy_high', 'dxy_low', 
              'gold_open', 'gold_close', 'gold_high', 'gold_low']

for col in float_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to float, coercing errors to NaN

# Check the data types to confirm changes
print(df.dtypes)
```
